In this notebook, I am aiming to investigate how well the number of attempts on an exercise is correlated with the amount of time it takes from the first submission to the last submission. I will have to filter out some of the data, as only attempts that eventually ended up being correct can be used. Furthermore, I will have to set a limit to the length of time (roughly say 2 hours) to filter out students who may have taken a break or continued the exercise at a later date. It's hard to parse this data out to identify how long it is taking students who are completing it in a single session, but I believe this is an okay way to do it.

A further analysis that would be cool to look at would be to identify the exercises where students took longer than 3 hours between their first attempt and correct attempt- suggesting potential difficulty.

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
# First separate date column into date and time
common_submissions = pd.read_csv('~/cisc5001/Common_Submissions.csv')
common_submissions['datetime'] = pd.to_datetime(common_submissions['date'], format='%m/%d/%y %H:%M')
common_submissions.head()
```

```python
# Filter submissions that eventually ended up being correct
correct_submissions = common_submissions.groupby(['user_id', 'exercise_ssn']).filter(lambda x: x['is_correct'].any())

# Calculate total time and number of attempts for each user-exercise pair
time_diff = correct_submissions.groupby(['user_id', 'exercise_ssn']).agg(
    first_submission=('datetime', 'min'),
    last_submission=('datetime', 'max'),
    attempts=('submission_id', 'count')
).reset_index()

# Calculate total time taken for each user-exercise pair
time_diff['total_time'] = (time_diff['last_submission'] - time_diff['first_submission']).dt.total_seconds() / 3600

# Filter for submissions within 2 hours
time_diff_filtered = time_diff[time_diff['total_time'] <= 2]
time_diff_filtered.head()
```

Let's plot this now and draw relationships!

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Plot
slope, intercept, r_value, p_value, std_err = linregress(time_diff_filtered['attempts'], time_diff_filtered['total_time'])
plt.figure(figsize=(10, 6))
sns.scatterplot(data=time_diff_filtered, x='attempts', y='total_time', color='blue', s=60, alpha=0.7)
sns.regplot(data=time_diff_filtered, x='attempts', y='total_time', color='red', scatter=False, label=f'R = {r_value**2:.4f}, p = {p_value:.4f}')
plt.title('Correlation between Attempts and Time Taken (2-hour limit)')
plt.xlabel('Number of Attempts')
plt.ylabel('Total Time (hours)')
plt.legend()
plt.show()
```

Correlation isn't very strong, but it does seem (obviously) that students are taking more time to submit more entries (within this 2 hour time limit). I wonder if we parse it out only by students who successfully complete the exercise in 2 hours.

```python
# Filter to keep only correct submissions and calculate total time within 2 hour limit
correct_submissions = common_submissions.groupby(['user_id', 'exercise_ssn']).filter(lambda x: x['is_correct'].any())
correct_submissions['datetime'] = pd.to_datetime(correct_submissions['datetime'])

time_diff = correct_submissions.groupby(['user_id', 'exercise_ssn']).agg(
    first_submission=('datetime', 'min'),
    last_submission=('datetime', 'max')
).reset_index()
time_diff['total_time_minutes'] = (time_diff['last_submission'] - time_diff['first_submission']).dt.total_seconds() / 60
time_diff = time_diff[time_diff['total_time_minutes'] <= 120]  

# Average time per exercise for students who completed it
avg_time_per_exercise = time_diff.groupby('exercise_ssn')['total_time_minutes'].mean().reset_index(name='avg_time_taken_minutes')

# Calculate the % of students who never answered correctly for each exercise
total_students_per_exercise = common_submissions.groupby('exercise_ssn')['user_id'].nunique()
students_never_correct = common_submissions.groupby(['exercise_ssn', 'user_id'])['is_correct'].any().groupby('exercise_ssn').apply(lambda x: x.eq(False).sum())

percent_never_correct = (students_never_correct / total_students_per_exercise * 100).reset_index(name='%_never_correct')
percent_never_correct.fillna(0, inplace=True)

# Merge datasets
exercise_analysis = pd.merge(avg_time_per_exercise, percent_never_correct, on='exercise_ssn')
```

```python
# Plot
slope, intercept, r_value, p_value, std_err = linregress(exercise_analysis['avg_time_taken_minutes'], exercise_analysis['%_never_correct'])
plt.figure(figsize=(10, 6))
sns.scatterplot(data=exercise_analysis, x='avg_time_taken_minutes', y='%_never_correct', color='blue', s=60, alpha=0.7)
sns.regplot(data=exercise_analysis, x='avg_time_taken_minutes', y='%_never_correct', color='red', scatter=False, label=f'R = {r_value**2:.4f}, p = {p_value:.4f}')
plt.title('Correlation between Avg Time Taken for Students to Successfully Complete an Exercise\n and % of Students Who Attempted an Exercise but Never Completed It')
plt.xlabel('Average Time Taken (minutes)')
plt.ylabel('% of Students Who Never Answered Correctly')
plt.legend()
plt.show()
```

It seems that the exercises that students took the longest to answer correctly also tended to be the exercises where students gave up the most on.

It seems that many of these metrics are correlated with each other and suggest that exercises with higher time taken, # of attempts, and % of students who gave up seem to be the more difficult exercises.

I am now curious about comparing the average amount of time it takes students to complete an exercise from the first attempt to the correct attempt versus the amount of time it takes for students from their first attempt to the last attempt before they inevitably give up and step away from their system (again, arbitrary 3 hour rule applies here). HOWEVER, the kicker here is that I am going to look at the top 17 "challenging" exercises that was determined by looking at the exercises with the most submissions! A welch's t-test will be used.

```python
# Filter the dataset for challenging exercises
challenging_exercises = pd.read_csv('~/cisc5001/top17challenging.csv')
challenging_submissions = common_submissions[common_submissions['exercise_ssn'].isin(challenging_exercises['exercise_ssn'])]

# Filter for correct and gave-up attempts
correct_submissions = challenging_submissions.groupby(['user_id', 'exercise_ssn']).filter(lambda x: x['is_correct'].any())
gave_up_submissions = challenging_submissions.groupby(['user_id', 'exercise_ssn']).filter(lambda x: not x['is_correct'].any())

# Correct and gave-up time calculations
correct_time = correct_submissions.groupby(['user_id', 'exercise_ssn']).agg(
    first_submission=('datetime', 'min'),
    correct_submission=('datetime', lambda x: x[correct_submissions['is_correct']].min())
).reset_index()
correct_time['time_to_correct'] = (correct_time['correct_submission'] - correct_time['first_submission']).dt.total_seconds() / 60

gave_up_time = gave_up_submissions.groupby(['user_id', 'exercise_ssn']).agg(
    first_submission=('datetime', 'min'),
    last_submission=('datetime', 'max')
).reset_index()
gave_up_time['time_to_give_up'] = (gave_up_time['last_submission'] - gave_up_time['first_submission']).dt.total_seconds() / 60

# Filter within 2 hour limit
correct_time_filtered = correct_time[correct_time['time_to_correct'] <= 120]
gave_up_time_filtered = gave_up_time[gave_up_time['time_to_give_up'] <= 120]

# Welch's t-test
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(correct_time_filtered['time_to_correct'], gave_up_time_filtered['time_to_give_up'], equal_var=False)
print(f"Welch's t-test: t-stat = {t_stat:.4f}, p-value = {p_value:.4f}")
```

```python
# Combine datasets with appropriate labels for plotting
correct_time_filtered['label'] = 'Time to Correct'
gave_up_time_filtered['label'] = 'Time to Give Up'

# Combine both datasets for plotting
combined_time = pd.concat([
    correct_time_filtered[['time_to_correct', 'label']].rename(columns={'time_to_correct': 'time'}),
    gave_up_time_filtered[['time_to_give_up', 'label']].rename(columns={'time_to_give_up': 'time'})
])

# Plotting boxplot for comparison
plt.figure(figsize=(10, 6))
sns.boxplot(data=combined_time, x='label', y='time', palette='pastel', showmeans=True)
plt.title('Comparison of Time Taken: Correct vs Gave Up (Filtered within 2 hours)')
plt.xlabel('Outcome')
plt.ylabel('Time (minutes)')

# Welch's t-test
t_stat, p_value = ttest_ind(
    correct_time_filtered['time_to_correct'], 
    gave_up_time_filtered['time_to_give_up'], 
    equal_var=False
)

# Display the p-value on the plot
p_value_text = f"p = {p_value:.4f}"
y_max = combined_time['time'].max()
plt.text(0.5, y_max * 0.95, p_value_text, ha='center', fontsize=12, bbox=dict(facecolor='white', alpha=0.6))
plt.show()
```

It seems that students generally tend to spend more time on these 17 difficult problems before they give up than their fellow peers who did not give up and got the question right (within 2 hours). I am curious to uncap the time now and look at this same analysis but without the time capped.

Ok great, same trend here. How about calculating the number of attempts on these exercises before getting it correct vs the number of attempts before students give up?

```python
# Group by user and exercise, then count attempts
attempts_completed = correct_submissions.groupby(['user_id', 'exercise_ssn']).size().reset_index(name='attempts')
attempts_gave_up = gave_up_submissions.groupby(['user_id', 'exercise_ssn']).size().reset_index(name='attempts')

# Remove outliers from both groups using IQR
filtered_attempts_completed = remove_outliers(attempts_completed, 'attempts').assign(Outcome='Completed')
filtered_attempts_gave_up = remove_outliers(attempts_gave_up, 'attempts').assign(Outcome='Gave Up')

# Combine filtered datasets
filtered_combined_attempts = pd.concat([filtered_attempts_completed, filtered_attempts_gave_up])

# Welch's t-test
t_stat_attempts, p_value_attempts = ttest_ind(
    filtered_attempts_completed['attempts'], 
    filtered_attempts_gave_up['attempts'], 
    equal_var=False
)

# Plot the boxplot for number of attempts
plt.figure(figsize=(10, 6))
sns.boxplot(data=filtered_combined_attempts, x='Outcome', y='attempts', palette='pastel', showmeans= True)
plt.title('Comparison of Number of Attempts to Complete vs. Give Up')
plt.xlabel('Outcome')
plt.ylabel('Number of Attempts')
p_value_text_attempts = f"p = {p_value_attempts:.4f}"
plt.text(0.5, filtered_combined_attempts['attempts'].max() * 0.95, 
         p_value_text_attempts, ha='center', fontsize=12, bbox=dict(facecolor='white', alpha=0.6))

plt.show()
print(p_value_attempts)
```
